#### Chart - 1

````python
# Chart - 1: Histogram showing frequency distribution of prices of rooms

# Setting the size of the visual
fig = plt.figure(figsize=(24, 8)) 

# Plotting the histogram
ax = df['price'].plot(kind='hist',bins=500,edgecolor='black')

# Changing the range of x-axis in the histogram
ax.set_xlim([0,1000])
plt.xticks([0,100,200,300,400,500,600,700,800,900,1000])

# Setting the axis titles and the graph title
plt.xlabel('Price')
plt.title('Distribution of Price')
````

1. Why did you pick the specific chart?

In the case of room prices, the histogram can help us understand how the prices are distributed across a range of values. We can see how many rooms fall into different price ranges and identify any patterns or trends in the data, such as a skewness towards higher or lower prices.

Other chart types, such as boxplots or scatterplots, can also be used to visualize the distribution of continuous data. However, a histogram is particularly useful when we want to explore the shape of the distribution and the frequency of observations within different ranges. It can also help us identify potential outliers or unusual values in the data.

##### 2. What is/are the insight(s) found from the chart?

From the histogram, we can infer the following insights about the distribution of room prices:

Skewness: The distribution of room prices is right-skewed, which means that there are more listings with lower prices and fewer listings with higher prices. This is evident from the fact that the histogram bars are taller on the left side and shorter on the right side.

Central Tendency: The mean price of listings is 152.72, which is higher than the median price of 106. This indicates that the distribution is skewed towards higher prices, and there are a few listings with very high prices that are pulling the mean upwards.

Outliers: The histogram shows that there are very few listings with prices higher than 300, and almost none with prices higher than 500. This suggests that there may be some outliers or extreme values in the dataset that are influencing the overall distribution.

````python
# Chart - 7: Variation in the number of bookings day wise and month wise

# Converting Last Review column to datetime datatype
df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')
df['Days'], df['Month'], df['Years'] = (df['last_review'].dt.day, df['last_review'].dt.month, df['last_review'].dt.year)
df['last_review'] = pd.to_datetime(df['last_review']).dt.date

# Null column date becomes system min date i.e. 1970-01-01 so we're filtering out these datas
filtered_df = df[df['Years'] != 1970]

# Plotting the results
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(24, 8))
ax = axes.flatten()

sns.histplot(data=filtered_df, x='Days', hue='room_type', multiple="stack", ax=ax[0])
ax[0].set_title('No of Guest in days of the months')
sns.histplot(data=filtered_df, x='Month', hue='room_type', multiple="stack", ax=ax[1])
ax[1].set_title('No of Guest in months of the years')
sns.despine(fig, left=True)
````
